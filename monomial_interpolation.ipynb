```python
import numpy as np
import matplotlib.pyplot as plt
import math
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
```

```python
# calculate some px
def px(x, y):
    t = np.zeros((len(x), len(x)))
    for i in range(len(x)):
        for j in range(len(x)):
            t[i][j] = x[i]**j
    c = np.linalg.solve(t, y)
    return c
```

```python
# evalulates some px at given point x
def ev(pn, x):
    y = 0
    for i in range(len(pn)):
        y += (pn[i]*(x**i))
    return y
```

```python
# evaluvating f at given points
f = lambda x : np.sin(x)
x = [0, np.pi/6, np.pi/4, np.pi/3, np.pi/2]
y = [f(i) for i in x]
p4 = px(x, y)
```

```python
# error
ev(p4, 1.2) - f(1.2)
```

```python
# points for graphing
p4x = [i for i in np.arange (0, np.pi/2, 0.1)]
p4y = [ev(p4, i) for i in np.arange (0, np.pi/2, 0.1)]
```

```python
# sin(x) at 5 points
plt.plot(x, y, color='blue')
# interpolating polynomial
plt.plot(p4x, p4y, color='red')
```
